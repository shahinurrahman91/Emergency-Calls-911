```python
# Import numpy and pandas
```

```python
import numpy as np
import pandas as pd
import datetime
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import visualization libraries and set %matplotlib inline
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
dateparse = lambda x: datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S')
```

```python
# Read the dataset
```

```python
df=pd.read_csv("911.csv",
    header=0,names=['lat', 'lng','desc','zip','title','timeStamp','twp','addr','e'],
    dtype={'lat':str,'lng':str,'desc':str,'zip':str,
                  'title':str,'timeStamp':str,'twp':str,'addr':str,'e':int}, 
     parse_dates=['timeStamp'],date_parser=dateparse)
```

```python
# Set index
```

```python
df.index = pd.DatetimeIndex(df.timeStamp)
df=df[(df.timeStamp >= "2016-01-01 00:00:00")]
```

```python
df
```

```python
df.head()
```

```python
df['twp'].value_counts().head(5)
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
# Differentiating reasons for an emergency call
```

```python
sns.countplot(x='Reason',data=df,palette='viridis')
```

```python
# Use pd.to_datetime to convert the column from strings to DateTime objects 
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
# Creating three columns of hour, month and date
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
# Use of the .map() with this dictionary to map the actual string names to the day of the week
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
# Now, we use seaborn to create a countplot of the Day of Week column 
```

```python
sns.countplot(x='Day of Week',data=df,hue='Reason',palette='viridis')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# Repeat the same procedure for month
```

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.countplot(x='Hour',data=df,hue='Reason',palette='viridis')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# create a pivot table with only EMS
```

```python
r=df[df['Reason'] == 'EMS' ]
t=pd.pivot_table(r, values='e', index=['timeStamp'], columns=['title'], aggfunc=np.sum)
```

```python
# Resample by every week 'W'
```

```python
pt=t.resample('W', how=[np.sum]).reset_index()
pt.head()
```

```python
# 'sum' column is not necessary
```

```python
pt.columns = pt.columns.get_level_values(0)
pt.head()
```

```python
# category of the call
df["title"].value_counts()
```

```python
# Line showed with the green dot
fig, av = plt.subplots()

av.plot_date(pt['timeStamp'], pt['EMS: ASSAULT VICTIM'],'k')
av.plot_date(pt['timeStamp'], pt['EMS: ASSAULT VICTIM'],'go')

av.set_title("EMS: ASSAULT VICTIM")
fig.autofmt_xdate()
plt.show()
```
